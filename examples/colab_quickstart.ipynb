```python
import pyield as yd
from pyield import bday

print(yd.__version__)
```

## 1. Dias Úteis (Calendário)
A função `bday.count` calcula o número de dias úteis entre duas datas. Formatos aceitos: `DD-MM-YYYY` ou `YYYY-MM-DD`.

```python
dias = bday.count("02-01-2024", "09-01-2024")
print("Dias úteis:", dias)
# Lista de vários períodos (retorna Series Polars)
periodos = bday.count("01-01-2024", ["08-01-2024", "22-01-2024"])
print(periodos)
```

### Offset de data útil
`bday.offset(data, n)` ajusta para o próximo dia útil considerando feriados.

```python
ajustada = bday.offset("20-11-2024", 0)  # 20/11/2024 é feriado nacional recente
print("Data ajustada:", ajustada)
```

## 2. Futuros DI (B3)
Obtém dados de contratos futuros de DI1 (taxas de vencimento). Faz requisição à B3, portanto é necessário acesso à internet.

```python
df = yd.futures(contract_code="DI1", date="26-12-2023")
df.head()
```

Converter para Pandas (se desejar interoperar com outras libs):

```python
df_pandas = df.to_pandas()
df_pandas.head()
```
